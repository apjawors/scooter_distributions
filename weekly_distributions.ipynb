```python
import pandas as pd
import numpy as np
import operator
from itertools import islice
```

```python
#Files to name and read
zones = pd.read_csv('zones.csv')
trips = pd.read_csv('trips.csv')
bird_map = pd.read_csv('bird.csv', skiprows=1)
lime_map = pd.read_csv('lime.csv', skiprows=1)
#spin_map = pd.read_csv('spin.csv', skiprows=1)

#Create pivot table of zones.csv for readability
table = pd.pivot_table(zones, values='average_total_count', index='date', columns=['operator', 'shape_name'])
```

```python
#Insert 'Bird Total' and 'Lime Total' columns to table
table['Bird Total'] = table.iloc[:, 0:4].sum(axis=1)
table['Lime Total'] = table.iloc[:, 4:8].sum(axis=1)
#table['Spin Total'] = table.iloc[:, 8:12].sum(axis=1)

#Get total operator vehicles in access zones for the week and total operator vehicles overall 
birdTotal = table['Bird Total'].sum()
birdAZTotal = table['bird']['Access Zone'].sum()

limeTotal = table['Lime Total'].sum()
limeAZTotal = table['lime']['Access Zone'].sum()

#spinTotal = table['Spin Total'].sum()
#spinAZTotal = table['spin']['Access Zone'].sum()

table
```

```python
#Get total operator trips for the week
totalBirdTrips = trips['bird'].sum()
totalLimeTrips = trips['lime'].sum()
#totalSpinTrips = trips['spin'].sum()

#Calculate average vehicle count for each operator
birdAvgVehicles = round(table['Bird Total'].mean(), 2)
limeAvgVehicles = round(table['Lime Total'].mean(), 2)
#spinAvgVehicles = round(table['Spin Total'].mean(), 2)

#Calculate MUR for each operator
birdMUR = round(totalBirdTrips / (birdTotal - birdAZTotal), 2)
limeMUR = round(totalLimeTrips / (limeTotal - limeAZTotal), 2)
#spinMUR = round(totalSpinTrips / (spinTotal - spinAZTotal), 2)

#Get AZ% and HUZ% for each company from 'lime', 'bird', and 'spin' files
birdAZ = round(bird_map['percent'][0], 2)
birdHUZ = round(bird_map['percent'][1], 2)

limeAZ = round(lime_map['percent'][0], 2)
limeHUZ = round(lime_map['percent'][1], 2)

#spinAZ = round(spin_map['percent'][0], 2)
#spinHUZ = round(spin_map['percent'][1], 2)

#Create dataframe out of dictionary of distributions
d = {'': ['Bird', 'Lime', 'Spin'], 
     'Vehicle Count': [birdAvgVehicles, limeAvgVehicles, """spinAvgVehicles"""], 
     'MUR': [birdMUR, limeMUR, """spinMUR"""], 
     'AZ%': [birdAZ, limeAZ, """spinAZ"""], 
     'HUZ%': [birdHUZ, limeHUZ, """spinHUZ"""]}

distributions = pd.DataFrame(data=d).set_index('')

#Final result
distributions
```
